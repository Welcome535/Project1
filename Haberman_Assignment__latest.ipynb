```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

'''Download Haberman Dataset'''
haberman= pd.read_csv("haberman.CSV")
haberman.head(8)

```

```python
'''Data Points'''
print('Dataset shape is:', haberman.shape)
```

```python

'''columns Names'''
print(haberman.columns)
```

```python
'''Finding the data points'''
haberman['Surv_status'].value_counts()
```

```python
observation:
    
    1)There are 306 total data points.
    2)out of which 225 Data points are survival status of maore than 5 years.
    3)81 patients are not survives more than 5 five years .this is imbalanced dataset.
    
```

```python
'''2-D Scatter Plot'''
haberman.plot(kind = 'scatter' , y='axil_nodes' , x='Age')
plt.grid()
plt.show()
```

```python
observation: if the no of axillary nodes are detected more than 15, they belong to Age group in 35 to 70 group of age.
```

```python
'''2-D Scatter plot with color-coding for each class.'''
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="Surv_status", size=5) \
   .map(plt.scatter, "Age", "axil_nodes") \
   .add_legend();
plt.show();
```

```python
observation: Blue points denotes the patients survied more than 5 years and orage denotes to not survival of more than 5 years and also shows the patients weith zero axillary points cna also survival of more than 5 five years.
```

```python
'''Pair-plot'''
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman, hue="Surv_status",vars = ['Age', 'Op_Year', 'axil_nodes'], size=3)\
.add_legend()
plt.show()
```

```python
observation: Due to overlap not able to classify which is the most efficient feature.hence we dont get much infomation from pair plots. 
```

```python
sns.FacetGrid(haberman, hue="Surv_status", size=4) \
   .map(sns.distplot, "axil_nodes") \
   .add_legend();
plt.show();
```

```python
observation: there is majaor overlap of axillary nodes. overlap of 0 to 5 has more survival chances.
```

```python
'''univarient Ananlysis on op_year'''

sns.FacetGrid(haberman, hue="Surv_status", size=5) \
   .map(sns.distplot, "Op_Year") \
   .add_legend();
plt.show();
```

```python
observations:1) Based on the survival status from the operation year,there is majaor overlap of the data point.
             2) How ever from this graph we can observe there is more chance of patients survival from the operation year 1958 - 1963  .
```

```python
'''univarient Ananlysis on Age'''

sns.FacetGrid(haberman, hue="Surv_status", size=5) \
   .map(sns.distplot, "Age") \
   .add_legend();
plt.show();
```

```python
observation: 1)from this graph again we have more overlap over the data points.we dont get enough information from this.
             2)patients with age more than 30 and less than 40 has the higher chance of survival.
             3)patients with age more than 78 is not survived.
```

```python
'''PDF , CDF on operation year'''

counts, bin_edges = np.histogram(haberman['Op_Year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("PDF:" ,pdf);
print("Bin Edges :" ,bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show();
```

```python
observation: 1) this shows the PDF and CDF on operation year.
             2)based on the observation from the year 1930 to 1978 there is 15% to 20% of survival rate. 
```

```python
'''plotting PDF ,CDf for Age'''


counts, bin_edges = np.histogram(haberman['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("PDF:" ,pdf);
print("Bin Edges :" ,bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show();
```

```python
observation: Based on the above graph ,Age less than 60 has high chance of survival rate(19% of survival)
```

```python
'''box plot'''
sns.boxplot(x='Surv_status',y='axil_nodes', data=haberman)
plt.show()
```
